```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3
```

```python
RWC = 1
RW = 5 - RWC
RC = 3 -RWC
WC =  2 -RWC
R = 12 - RW - RC - RWC
W = 8 - RW -WC - RWC
C = 4 - RC -WC -  RWC
print("Дощових/холодних/вітряних-", RWC)
print("Дощових/вітряних-", RW)
print("Дощових/холодних-", RC)
print("Холодних/вітряних-", WC)
print("Дощових-", R)
print("Холодних-", C)
print("Вітряних-", W)
```

```python
venn3(subsets = (R, C, W, RC, WC, RW,   RWC ), set_labels = ('Дощових','Вітряних', 'Холодних'))
plt.show()
```
