```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3
```

```python
M_F_CH = 4
print('Математика/Фізика/Хімія',M_F_CH)
F_CH = 6 - M_F_CH
print('Фізика/Хімія',F_CH)
M_F = 7 - M_F_CH
print('Математика/Фізика',M_F)
M_CH = 9 - M_F_CH
print('Математика/Хімія',M_CH)
CH = 19 - F_CH - M_CH -M_F_CH
print('Хімія',CH)
M = 15 - M_F - M_CH - M_F_CH
print('Математика',M)
F = 16 - F_CH - M_F - M_F_CH
print('Фізика',F)


```

```python
venn3(subsets = (M, F, M_F, CH,M_CH, F_CH,M_F_CH), set_labels = ('Математика','Фізика', 'Хімія'))
plt.show()
```
