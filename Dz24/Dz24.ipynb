```python
# Задача 2
import math

X = [[1000, 1500, 2000], [0.5, 0.3, 0.2]]
Y = [[1000, 1500, 1750], [0.4, 0.4, 0.2]]

M2, M1, D1, D2 = 0,0,0,0

for i in range(len(X[0])):
         M1 = M1 + X[0][i]*X[1][i]
print("Мат.сподівання X--",M1)

for i in range(len(Y[0])):
         M2 = M2 + Y[0][i]*Y[1][i]
print("Мат.сподівання Y--", M2)

# в якості величини ступеня ризику принято використовувати дисперсію
for i in range(len(X[0])):
        D1 =(((X[0][i] - M1)*(X[0][i] - M1)) * X[1][i]) + D1
print("Дисперсія X ------", D1)

for j in range(len(Y[0])):
        D2 =(((Y[0][j] - M1)*(Y[0][j] - M1)) * Y[1][j]) + D2
print("Дисперсія Y ------", D2)

if D1<D2:
    print("Обираємо товар X з оцінки дисперсії")
else:   print("Обираємо товар Y  з оцінки дисперсії")
    
# додатково можна в якості оцінки ризику взяти середньоквадратичне відхилення
SV1 = math.sqrt(D1)
SV2 = math.sqrt(D2)

if SV1<SV2:
    print("Обираємо товар X з оцінки  середньоквадратичного відхилення")
else:   print("Обираємо товар Y з оцінки  середньоквадратичного відхилення")

```
